# Exercice 1 : Encapsulation

Le principe d'**encapsulation** est fondamental en programmation orientée objet. Il permet de regrouper, et éventuellement de masquer, des données derrière des outils qui facilitent ou généralisent leur manipulation.

```python
class Graphe:
    def __init__(self):
        self.__G = dict()
    
    def __len__(self):
        return len(self.__G)
                   
    def __str__(self):
        ch=""
        for v in self.__G:
            ch += str(v) + ":" + str(self.__G[v]) + '\n' # \n permet un retour a la ligne
        return ch
    #
    
    # Permet la syntaxe "for element in graphe[sommet]" 
    def __getitem__(self, v):
        return self.__G[v]

    # Permet la syntaxe G[v] = neighbors
    def __setitem__(self, v, neighbors):
        self.__G[v] = neighbors

    # Permet la syntaxe "for element in graphe" 
    def __contains__(self, v):
        return v in self.__G
    
    # ceci permet de pouvoir chercher directement les elem de G 
        
    def __iter__(self):
        return iter(self.__G)
    
    def max_degre(self):
        return max([len(self.__G[v]) for v in self.__G])
```

```python
G = Graphe()

G['v1'] = ['v2','v3']
G['v2'] = ['v4']
G['v3'] = ['v2']
G['v4'] = ['v5', 'v3']
G['v5'] = ['v4']

print(G)
print(len(G))
print("Degré de G:", G.max_degre())
#print(G.__G) # ici attribut privé donc ne peut pas fonctionner car en dehors de la classe
```

# Exercice 2 : Modélisation

L’objectif de cet exercice est de modéliser le labyrinthe ci-après sous la forme d'un graphe, sachant que : 
* Chaque case vide représente un sommet
* Deux sommets sont voisins si les cases sont adjacentes (haut/bas/gauche/droite)
* Un mur █ n'est donc pas un sommet.

```python
labyrinthe = """\
███████████████████████████
█                         █
█ █████  ████████  █ ███  █
█     █  █     █       █  █
█  █  █      █     █      █
█       ██  █     ████   ██
████      █  █            █
█       █ █    █████   ████
█  █    █   ████      █   █
█ ██    ███ ██ █   ███    █
█ █████  █     ██    ██ ███
█           █             █
███████████████████████████"""
```

Le labyrinthe est donné sous forme de chaîne de caractères. On peut le transformer en liste de listes, représentant chaque ligne du labyrinthe, comme suit :

```python
L = [list(line) for line in labyrinthe.splitlines()]
print(L)
```

</div>

1. Ecrire une fonction `contenu_case` qui retourne ce que contient une case donnée.

```python
def contenu_case(lab, case):
    return lab[case[0]][case[1]]
        

case=contenu_case(L, (10, 5))
print(case)
print(contenu_case(L, (1, 14)))
print(case)
print(contenu_case(L, (4, 7)))
```

2. Ecrire une fonction `voisins_accessibles` qui pour une case donnée liste les cases voisines qui sont accessibles (qui ne sont pas un mur).

```python
def voisins_accessibles(lab, case):
    voisins=[]
    offset = [(-1,0),(0,-1),(1,0),(0,1)]
    x,y = case
    nb_lignes = len(lab)
    nb_colonnes = len(lab[0])
    for dx,dy in offset: 
        if 0<= x + dx < nb_lignes and 0<= y + dy < nb_colonnes and contenu_case(lab,(x+dx,y+dy)) != '█':
            
        #x = position initiale de case , dx = position x de case suivante parmi offset
            voisins.append((x+dx, y+dy))
    return voisins

print(voisins_accessibles(L, (1, 14)))
print(voisins_accessibles(L, (6, 5)))
print(voisins_accessibles(L, (8, 8)))
```

3. Ecrire une fonction `construire_graphe`qui crée un objet `Graphe`correspondant au labyrinthe ci-dessus.

```python
def construire_graphe(lab):    
    G=Graphe()
    nb_lignes=len(lab)
    nb_colonnes=len(lab)
    for x in range(nb_lignes):
        for y in range(nb_colonnes):
            if contenu_case(lab,(x,y)) != '█':
                G[(x,y)]=voisins_accessibles(lab,(x,y))
    return G 
            

G_lab = construire_graphe(L)
print(G_lab)
```

## À la recherche de l’étoile perdue

Voici un nouveau labyrinthe contenant des symboles supplémentaires. Le symbole `*` est présent *exactement* une fois dans le labyrinthe, les symboles `o` apparaissant eux plusieurs fois. 

```python
labyrinthe2 = """\
███████████████████████████
█   o                  o  █
█ █████  ████████  █ ███  █
█     █  █     █       █  █
█  █  █      █     █      █
█       ██  █     ████   ██
████      █  █ *          █
█       █ █    █████   ████
█  █    █   ████      █   █
█ ██    ███ ██ █   ███    █
█ █████  █     ██    ██ ███
█   o       █           o █
███████████████████████████"""
```

1. Ecrire une fonction `coordonnees_symboles`qui renvoie les coordonnées du symbole `*` et la liste des coordonnées des symboles `o`.

```python
def coordonnees_symboles(lab):
    rond = []
    etoile = (0,0)
    for x in range(len(lab)):
        for y in range(len(lab[0])):
            if lab[x][y] == '*' :
                etoile = (x,y)
            elif lab[x][y] == 'o':
                rond.append((x,y))
    return etoile , rond

L2 = [list(line) for line in labyrinthe2.splitlines()]
coordonnees_symboles(L2)
```
